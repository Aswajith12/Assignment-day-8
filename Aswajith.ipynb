5. Handle outliers

```python
plt.boxplot(data['Age'])
plt.show()
```

```python
Q1 = np.percentile(data['Age'],25,interpolation='midpoint')
Q2 = np.percentile(data['Age'],50,interpolation='midpoint')
Q3 = np.percentile(data['Age'],75,interpolation='midpoint')
```

```python
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR = Q3-Q1
```

```python
print(IQR)
```

```python
up_lim = Q3+1.5*IQR
low_lim = Q1-1.5*IQR
```

```python
print(up_lim)
print(low_lim)
```

```python
outliers = []
for i in data['Age']:
    if i>up_lim or i<low_lim:
        outliers.append(i)
```
